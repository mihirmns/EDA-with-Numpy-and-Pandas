```python
##Q5.From the sales_data.csv file. Calculate the total sales price by the customer and draw the graph that shows the company on the x-axis and total sales on the y-axis. 
import socket
import datetime as dt
print("Mihir Nevpurkar , Laptop:{0}, Time:{1}".format(socket.gethostname(), dt.datetime.now()))

unique_companies = np.array(df['name'].unique())
print(unique_companies)

#df_c1 = df[df["name"] == "Carroll PLC"]["net_price"].sum()
#print("\n First COmpany sale is :",df_c1,"\n")

sizeofList = len(unique_companies)
result= []
result_array=[]
i= 0

while i < sizeofList:
    result = df[df["name"] == unique_companies[i]]["net_price"].sum()
    result_array.append(result)
    i = i+1
    

print("\n",result_array,"\n")

company_total_sales = pd.DataFrame(list(zip(unique_companies,result_array )),
               columns =['Name', 'Total Sales'])

print(company_total_sales)

name = company_total_sales['Name']
total_sales = company_total_sales['Total Sales']
 
# Figure Size
fig = plt.figure(figsize =(10, 7))
 
# Horizontal Bar Plot
plt.bar(name, total_sales, label='Total Sales')
plt.tick_params(axis='x', rotation=90)
 
# Show Plot
plt.show()


```

```python
##Q7
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import socket
import datetime as dt
print("Mihir Nevpurkar , Laptop:{0}, Time:{1}".format(socket.gethostname(), dt.datetime.now()))


  ##a) Loading the data 
stud_details= pd.DataFrame(
    data={'First_names': [ 'Jason','Molly','Tina','Jake','Amy'],
          'Last Names ': ['Miller','Jackson','fey','albert','NA' ],
          'gender ': ['m','f','f','m','f' ],
          'Age': [8,20,22,19,17],
         'ACT': [29,32,31,35,31],
         'SAT': [1280,1420,1360,1520,1395]},
    columns=['First_names', 'Last Names ', 'gender ', 'Age','ACT','SAT'])
stud_details

```
